```python
np.random.seed(43)
data = {"Category": ["A", "A", 'A', 'B', 'B', 'C', 'C', 'D', 'E', 'L', 'E', 'K' ], 
        'target': np.random.choice(2, size = 12) }
```

```python
df_example = pd.DataFrame.from_dict(data)
df_example
```

```python
ohe = OneHotEncoder(cols=['Category'])
result = ohe.fit_transform(df_example)
pd.concat([df_example, result], axis = 1)
```

```python
sum_en = SumEncoder()
result_sum = sum_en.fit_transform(df_example)
pd.concat([df_example, result_sum], axis = 1)
```

```python
helmert_enc = HelmertEncoder(verbose = True)
result_helmert = helmert_enc.fit_transform(df_example)
pd.concat([df_example, result_helmert], axis = 1)
```

```python
np.random.seed(43)
data = {"Category": ["A", "A", 'A', 'B', 'B', 'C', 'D'], 
        'target': np.random.choice(2, size = 7) }

df_example = pd.DataFrame.from_dict(data)
df_example
```

```python
target_enc = TargetEncoder()
result_target = target_enc.fit_transform(data['Category'], data['target'])
pd.concat([df_example, result_target], axis = 1)
```

```python
df_example.groupby('Category')['target'].mean()
```
